**1. Функція з параметром**

Створіть функцію `show_menu()`, яка приймає перелік страв, наприклад ['Піца', 'Паста', 'Салат'] і виводить меню ресторану в наступному вигляді:
```
=== МЕНЮ ===
1. Піца
2. Паста
3. Салат
=============
```

Викличте цю функцію на списку ["Суп грибний", "Котлета по-київськи", "Лимонад"].

````python
def show_menu(menu):
    print("=== МЕНЮ ===")
    for i, dish in enumerate(menu):
        print(f"{i+1}. {dish}")
    print("=============")

rest_menu = ["Суп грибний", "Котлета по-київськи", "Лимонад"]
show_menu(rest_menu)
````

**2. Функція з параметром**

Створіть функцію `print_square(n)`, яка виводить квадрат числа.

Наприклад: `print_square(5)` → "Квадрат числа 5 = 25"

Викличте її з різними числами.

````python
def print_square(n):
    print(f"Квадрат числа {n} = {n*n}")

print_square(3)
print_square(5)
print_square(8)
````

**3. Функція з поверненням значення**

Створіть функцію `multiply(a, b)`, яка повертає добуток двох чисел.

Викличте функцію на власноруч заданих значеннях та виведіть результат.


````python
def multiply(a, b):
    return a*b

res = multiply(3, 5)
print(res)
res = multiply(-6,-7)
print(res)
print(multiply(4, 5))
````

**4. Функція пошуку максимумального значення**

Створіть функцію `find_max(numbers)`, яка:
- Приймає список чисел
- Повертає найбільше число в списку

Протестуйте на списку `[15, 42, 8, 23, 56, 19]`.


````python
def find_max(numbers):
    return max(numbers)

test = [15, 42, 8, 23, 56, 19]
print(find_max(test))
````

**5. Функція з умовами**

Створіть функцію `check_grade(grade)`, яка:
- Приймає оцінку (число)
- Повертає "Здав" якщо grade >= 60
- Повертає "Не здав" якщо grade < 60

Протестуйте на різних оцінках.


````python
def check_grade(grade):
    if grade >= 60:
        return "Здав"
    else:
        return "Не здав"

print(check_grade(60))
print(check_grade(90))
print(check_grade(55))
````
