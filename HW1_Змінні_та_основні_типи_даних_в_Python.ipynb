**2. Арифметичні операції**

Створіть дві змінні: `var1` зі значенням 15, `var2` зі значенням 4.

Обчисліть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Суму
- Різницю
- Добуток
- Ділення
- Цілочисельне ділення
- Залишок від ділення
- Піднесення до степеня


```python
var1 = 15
var2 = 4
print(f"Sum: {var1} + {var2} = {var1 + var2}")
print(f"Difference: {var1} - {var2} = {var1 - var2}")
print(f"Product: {var1} * {var2} = {var1 * var2}")
print(f"Division: {var1} / {var2} = {var1 / var2}")
print(f"Integer division: {var1} // {var2} = {var1 // var2}")
print(f"Remainder: {var1} % {var2} = {var1 % var2}")
print(f"Exponentiation: {var1} ** {var2} = {var1 ** var2}") 
```
