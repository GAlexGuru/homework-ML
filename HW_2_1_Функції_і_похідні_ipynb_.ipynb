2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import numpy as np

x = sp.symbols('x')
f_sin = sp.sin(x)
f_ln = sp.ln(x)
print(f"Похідна sin(x): {sp.diff(f_sin, x)}")
print(f"Похідна ln(x): {sp.diff(f_ln, x)}")

x_sin = np.linspace(-2*np.pi, 2*np.pi, 400)
x_ln = np.linspace(0.1, 5, 400)

plt.subplot(2, 1, 1)
plt.plot(x_sin, np.sin(x_sin), label='sin(x)')
plt.plot(x_sin, np.cos(x_sin), label='cos(x)')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x_ln, np.log(x_ln), label='ln(x)')
plt.plot(x_ln, 1/x_ln, label='1/x')
plt.legend()
```
